```python
from collections import defaultdict
lineups_df = combined_lineups.copy()
players_lookup = defaultdict(set)

for _, row in lineups_df.iterrows():
    season = row["SEASON"]
    team = row["TEAM_ID"]
    pids = row["GROUP_ID"].split("-")
    for pid in pids:
        players_lookup[(season, team)].add(int(pid))
```

```python
def build_lineup_intervals(rot):
    intervals = {}

    for team_id, grp in rot.groupby("TEAM_ID"):
        # Each row is one player's stint
        # We need overlapping stints of 5 players

        # Build events
        events = []
        for r in grp.itertuples():
            events.append((r.IN_TIME_REAL, +1, r.PERSON_ID))
            events.append((r.OUT_TIME_REAL, -1, r.PERSON_ID))

        events.sort()

        active = set()
        last_t = None
        team_intervals = []

        for t, typ, pid in events:
            if last_t is not None and len(active) == 5:
                team_intervals.append(
                    (last_t, t, frozenset(active))
                )

            if typ == +1:
                active.add(pid)
            else:
                active.discard(pid)

            last_t = t

        intervals[team_id] = team_intervals

    return intervals

def lookup_lineup(intervals, team_id, t):
    for start, end, lineup in intervals.get(team_id, []):
        if start <= t < end:
            return lineup
    return None
```

```python
def shot_to_game_time(row):
    period_start = (row["PERIOD"] - 1) * 12 * 60
    seconds_into_period = (
        12 * 60
        - (row["MINUTES_REMAINING"] * 60 + row["SECONDS_REMAINING"])
    )
    return int((period_start + seconds_into_period) * 10)

def lineup_group_id(player_ids):
    return "-".join(str(pid) for pid in sorted(player_ids))

def lineup_key(player_ids):
    return frozenset(int(pid) for pid in player_ids)

lineups_df["lineup_key"] = lineups_df["GROUP_ID"].apply(
    lambda s: frozenset(int(pid) for pid in s.split("-"))
)

valid_lineup_keys = set(lineups_df["lineup_key"])

key_to_group_id = dict(
    zip(lineups_df["lineup_key"], lineups_df["GROUP_ID"])
)
```

```python
import time

import time

def log(msg):
    print(f"[{time.strftime('%H:%M:%S')}] {msg}")

errors = {
    "rotation_fail": [],
    "no_valid_lineups_in_game": [],
}

```

```python
from nba_api.stats.endpoints import shotchartdetail
def get_team_season_shots(team_id, season):
    shots = shotchartdetail.ShotChartDetail(
        team_id=team_id,
        player_id=0,
        season_nullable=season,
        season_type_all_star="Regular Season",
        context_measure_simple="FGA",
    ).get_data_frames()[0]
    return shots
```

```python
from nba_api.stats.endpoints import gamerotation

def process_game_from_shots(
    game_id,
    game_shots,
    season,
    team_id,
    team_season_lineups,
    relevant_players,
):
    # 1. Shooter filter
    game_shots = game_shots[
        game_shots["PLAYER_ID"].isin(relevant_players)
    ]
    if game_shots.empty:
        return None

    # 2. Rotations
    try:
        home_rot, away_rot = gamerotation.GameRotation(
            game_id=game_id
        ).get_data_frames()
    except Exception:
        errors["rotation_fail"].append(game_id)
        return None

    rot = pd.concat([home_rot, away_rot], ignore_index=True)

    # 3. Shot times
    game_shots = game_shots.copy()
    game_shots["SHOT_TIME_REAL"] = game_shots.apply(
        shot_to_game_time, axis=1
    )

    # 4. Build lineup intervals (ONCE per game)
    intervals = build_lineup_intervals(rot)

    # 5. Assign lineup_key per shot
    game_shots["lineup_key"] = game_shots.apply(
        lambda r: lookup_lineup(
            intervals,
            r["TEAM_ID"],
            r["SHOT_TIME_REAL"]
        ),
        axis=1
    )

    # 6. Filter to *your* lineups
    mask = game_shots["lineup_key"].isin(team_season_lineups)
    if not mask.any():
        errors["no_valid_lineups_in_game"].append(game_id)
        return None

    game_shots = game_shots[mask]

    # 7. Map back to GROUP_ID
    game_shots["lineup_id"] = game_shots["lineup_key"].map(key_to_group_id)

    # 8. Add identifiers
    game_shots["SEASON"] = season
    game_shots["TEAM_ID"] = team_id

    return game_shots[
        [
            "SEASON", "TEAM_ID", "GAME_ID",
            "PLAYER_ID", "PLAYER_NAME",
            "lineup_id",
            "LOC_X", "LOC_Y",
            "SHOT_MADE_FLAG",
            "SHOT_DISTANCE",
            "SHOT_ZONE_BASIC",
            "SHOT_ZONE_AREA",
            "SHOT_ZONE_RANGE",
            "PERIOD",
            "MINUTES_REMAINING",
            "SECONDS_REMAINING",
        ]
    ]
```

```python
def process_team_season_from_shots(season, team_id):
    log(f"=== {season} | TEAM {team_id} ===")

    shots = get_team_season_shots(team_id, season)
    if shots.empty:
        log("No shots for team-season")
        return []

    relevant_players = players_lookup[(season, team_id)]
    if not relevant_players:
        log("No relevant players for team-season")
        return []

    team_season_lineups = set(
        lineups_df.loc[
            (lineups_df["SEASON"] == season) &
            (lineups_df["TEAM_ID"] == team_id),
            "lineup_key"
        ]
    )

    log(
        f"Lineups in DB for team-season: {len(team_season_lineups)} | "
        f"Shots: {len(shots)}"
    )

    if not team_season_lineups:
        log("No lineups in DB for this team-season")
        return []

    game_groups = dict(tuple(shots.groupby("GAME_ID")))
    out = []

    for game_id, game_shots in game_groups.items():
        df = process_game_from_shots(
            game_id,
            game_shots,
            season,
            team_id,
            team_season_lineups,
            relevant_players,
        )

        if df is not None:
            out.append(df)

        time.sleep(3.5)  # anti-block

    return out
```

```python
all_shots = []

for row in (
    lineups_df[["SEASON", "TEAM_ID"]]
    .drop_duplicates()
    .itertuples(index=False)
):
    dfs = process_team_season_from_shots(
        row.SEASON,
        row.TEAM_ID
    )
    if dfs:
        all_shots.extend(dfs)

shots_with_lineups = (
    pd.concat(all_shots, ignore_index=True)
    if all_shots else pd.DataFrame()
)
```

```python
import pandas as pd
import numpy as np
import os

# -----------------------------
# LOAD INPUT TABLES
# -----------------------------
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))

shots_preview = pd.read_csv(
    os.path.join(PROJECT_ROOT, "data", "interim", "lineups", "allstar_shots_with_lineups_preview.csv")
)

lineups_syn = pd.read_csv(
    os.path.join(PROJECT_ROOT, "data", "processed", "allstar_lineups_data_synthetic.csv")
)

# -------------------------------------------------
# FILTER LUKA COMBINATIONS
# -------------------------------------------------
STAR_NAME = "Luka Doncic"
STAR_ID = 1629029
TOTAL_SHOTS = 35_000

luka_combos = (
    lineups_syn[lineups_syn["star_player"] == STAR_NAME]
    .reset_index(drop=True)
)

# -------------------------------------------------
# BUILD REALISTIC (X, Y) DISTRIBUTIONS PER ZONE
# -------------------------------------------------
zone_xy_lookup = {}

group_cols = ["SHOT_ZONE_BASIC", "SHOT_ZONE_RANGE"]

for (basic, rng), grp in shots_preview.groupby(group_cols):
    coords = grp[["LOC_X", "LOC_Y"]].dropna()
    if len(coords) >= 20:  # avoid tiny groups
        zone_xy_lookup[(basic, rng)] = coords.reset_index(drop=True)

# Fallback pool (all shots)
fallback_xy = shots_preview[["LOC_X", "LOC_Y"]].dropna().reset_index(drop=True)

# -------------------------------------------------
# LEARN MAKE RATES
# -------------------------------------------------
zone_make_probs = (
    shots_preview
    .groupby("SHOT_ZONE_BASIC")["SHOT_MADE_FLAG"]
    .mean()
    .to_dict()
)

GLOBAL_MAKE_PROB = shots_preview["SHOT_MADE_FLAG"].mean()

# -------------------------------------------------
# RANDOM SHOT LOCATIONS (REALISTIC RANGES)
# -------------------------------------------------
def sample_xy(zone_basic, zone_range):
    key = (zone_basic, zone_range)

    if key in zone_xy_lookup:
        row = zone_xy_lookup[key].sample(1).iloc[0]
    else:
        row = fallback_xy.sample(1).iloc[0]

    # NBA shot locations are ints
    return int(row.LOC_X), int(row.LOC_Y)


# -------------------------------------------------
# SAMPLE SHOT CONTEXT (EXCEPT MAKE FLAG)
# -------------------------------------------------
shot_attr_cols = [
    "SHOT_DISTANCE",
    "SHOT_ZONE_BASIC",
    "SHOT_ZONE_AREA",
    "SHOT_ZONE_RANGE",
    "PERIOD",
    "MINUTES_REMAINING",
    "SECONDS_REMAINING",
]

shot_attrs = shots_preview[shot_attr_cols].sample(
    n=TOTAL_SHOTS,
    replace=True,
    random_state=42
).reset_index(drop=True)

# -------------------------------------------------
# GENERATE REALISTIC LOC_X / LOC_Y (INT)
# -------------------------------------------------
xy_samples = shot_attrs.apply(
    lambda r: sample_xy(
        r["SHOT_ZONE_BASIC"],
        r["SHOT_ZONE_RANGE"]
    ),
    axis=1
)

shot_attrs["LOC_X"] = [x for x, y in xy_samples]
shot_attrs["LOC_Y"] = [y for x, y in xy_samples]

# -------------------------------------------------
# GENERATE SHOT_MADE_FLAG
# -------------------------------------------------
shot_attrs["SHOT_MADE_FLAG"] = shot_attrs["SHOT_ZONE_BASIC"].apply(
    lambda z: int(np.random.rand() < zone_make_probs.get(z, GLOBAL_MAKE_PROB))
)

# -------------------------------------------------
# ASSIGN ARCHETYPE COMBINATIONS
# -------------------------------------------------
combo_idx = np.random.choice(
    luka_combos.index,
    size=TOTAL_SHOTS,
    replace=True
)

combo_rows = luka_combos.loc[combo_idx].reset_index(drop=True)

# -------------------------------------------------
# BUILD FINAL TABLE
# -------------------------------------------------
synthetic_luka_shots = pd.DataFrame({
    "PLAYER_ID": STAR_ID,
    "PLAYER_NAME": STAR_NAME,
    "star_player": STAR_NAME,
    "player1_archetype": combo_rows["player1_archetype"].values,
    "player2_archetype": combo_rows["player2_archetype"].values,
    "player3_archetype": combo_rows["player3_archetype"].values,
    "player4_archetype": combo_rows["player4_archetype"].values,
})

synthetic_luka_shots = pd.concat(
    [synthetic_luka_shots, shot_attrs],
    axis=1
)


# -------------------------------------------------
# SANITY CHECKS
# -------------------------------------------------
assert synthetic_luka_shots["SHOT_MADE_FLAG"].isin([0, 1]).all()
assert synthetic_luka_shots.isnull().sum().sum() == 0

print("Synthetic Luka shots (with realistic makes) ready:")
display(synthetic_luka_shots.head())
```

adding season column to all the allstar tables

```python
project_root = "../"
data_path = project_root + "data/raw/lineups/"
full_lineups_data = pd.read_csv(data_path + "allstar_lineups_full_team.csv")
advanced_lineups_data = pd.read_csv(data_path + "allstar_lineups_data_advanced.csv")
opponent_lineups_data = pd.read_csv(data_path + "allstar_lineups_data_opponent.csv")
scoring_lineups_data = pd.read_csv(data_path + "allstar_lineups_data_scoring.csv")
misc_lineups_data = pd.read_csv(data_path + "allstar_lineups_data_misc.csv")
base_lineups_data = pd.read_csv(data_path + "allstar_lineups_data_base.csv")
```

```python
# add SEASON from full_lineups_data to the other lineup tables by GROUP_ID + SUM_TIME_PLAYED

def normalize_keys(df):
    df = df.copy()
    if "GROUP_ID" in df.columns:
        df["GROUP_ID"] = df["GROUP_ID"].astype(str).str.strip()
    if "SUM_TIME_PLAYED" in df.columns:
        df["SUM_TIME_PLAYED"] = pd.to_numeric(df["SUM_TIME_PLAYED"], errors="coerce")
    return df

# prepare lookup (drop duplicates to avoid many-to-one issues)
full_lookup = normalize_keys(full_lineups_data)[["GROUP_ID", "SUM_TIME_PLAYED", "SEASON"]].drop_duplicates(
    subset=["GROUP_ID", "SUM_TIME_PLAYED"]
)

def add_season_column(df, lookup=full_lookup):
    df = normalize_keys(df)
    merged = df.merge(lookup, on=["GROUP_ID", "SUM_TIME_PLAYED"], how="left", validate="m:1")
    missing = merged["SEASON"].isna().sum()
    print(f"Added SEASON → missing: {missing} / {len(merged)}")
    return merged

advanced_lineups_data = add_season_column(advanced_lineups_data)
opponent_lineups_data = add_season_column(opponent_lineups_data)
scoring_lineups_data = add_season_column(scoring_lineups_data)
misc_lineups_data = add_season_column(misc_lineups_data)
base_lineups_data = add_season_column(base_lineups_data)

# optionally persist updated tables
data_path = project_root + "data/interim/lineups/"
advanced_lineups_data.to_csv(data_path + "allstar_lineups_data_advanced_with_season.csv", index=False)
opponent_lineups_data.to_csv(data_path + "allstar_lineups_data_opponent_with_season.csv", index=False)
scoring_lineups_data.to_csv(data_path + "allstar_lineups_data_scoring_with_season.csv", index=False)
misc_lineups_data.to_csv(data_path + "allstar_lineups_data_misc_with_season.csv", index=False)
base_lineups_data.to_csv(data_path + "allstar_lineups_data_base_with_season.csv", index=False)
```
