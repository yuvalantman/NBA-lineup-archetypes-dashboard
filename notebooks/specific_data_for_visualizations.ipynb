```python
import os
import pandas as pd
import numpy as np
```

```python
allstar_ids = [203999, 201939, 201935, 202695, 2544, 1629029, 203507, 1630162, 1626164, 201142, 1628983]

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
raw_path = os.path.join(PROJECT_ROOT, "data", "raw", "clusters", "General_player_features.csv")
general_player_data = pd.read_csv(raw_path)
allstar_data = general_player_data[general_player_data['PLAYER_ID'].isin(allstar_ids)].drop(columns=["Unnamed: 0"]).reset_index(drop=True)
allstar_data
```

```python
current_team = {
    203999: "Denver Nuggets",
    201939: "Golden State Warriors",
    201935: "Los Angeles Clippers",
    202695: "Los Angeles Clippers",
    2544: "Los Angeles Lakers",
    1629029: "Los Angeles Lakers",
    203507: "Milwaukee Bucks",
    1630162: "Minnesota Timberwolves",
    1626164: "Phoenix Suns",
    201142: "Houston Rockets",
    1628983: "Oklahoma City Thunder"}
allstar_data['CURRENT_TEAM'] = allstar_data['PLAYER_ID'].map(current_team)
allstar_data.to_csv(os.path.join(PROJECT_ROOT, "data", "processed", "allstar_data.csv"), index=False)
```

```python
luka_lineups = pd.read_csv(os.path.join(PROJECT_ROOT, "data", "processed", "allstar_lineups_data_synthetic.csv"))
luka_lineups
```

```python
efficiency_graph = luka_lineups[['star_player','player1_archetype','player2_archetype','player3_archetype', 'player4_archetype', 'offensive_rating', 'defensive_rating', 'net_rating']]
efficiency_graph.to_csv(os.path.join(PROJECT_ROOT, "data", "processed", "luka_efficiency_graph_data.csv"), index=False)
```
